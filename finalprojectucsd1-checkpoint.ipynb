```python
correlations=data.corr()
attrs = correlations.iloc[:-1,:-1] # all except target
threshold = 0.5
important_corrs = (attrs[abs(attrs) > threshold][attrs != 1.0]) \
.unstack().dropna().to_dict()
unique_important_corrs = pd.DataFrame(
list(set([(tuple(sorted(key)), important_corrs[key]) \
for key in important_corrs])),
columns=['Attribute Pair', 'Correlation'])
# sorted by absolute value
unique_important_corrs = unique_important_corrs.loc[
abs(unique_important_corrs['Correlation']).argsort()[::-1]]
plt.figure(dpi = 200)
unique_important_corrs.plot(x = 'Attribute Pair', y = 'Correlation',kind = 'barh',figsize =(15,10))
```

```python
cols_miss = ['ReasonsForLack']
data = data.drop(cols_miss, axis = 1)
```

```python
data.isna().sum()
```

```python
data.nunique
```

```python
cols_le = ['EducationSector','KeyTraits']
cols_oh = ['Gender','City','Influenced','MentalDisorder']
```

```python
data['KeyTraits'].unique()
```

```python
LE = LabelEncoder()
data[cols_oh] = data[cols_oh].apply(LE.fit_transform)
```

```python
data['IndividualProject'] = LE.fit_transform(data['IndividualProject'])
```

```python
data[cols_le] = data[cols_le].astype(str)
data = pd.get_dummies(data , columns = cols_le)
```

```python
data[cols_oh]
```

```python
X = data.drop('y',axis = 1)
```

```python
y = data.y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
X_train.keys()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[['Age','Perseverance','DesireToTakeInitiative','Competitiveness','SelfReliance','StrongNeedToAchieve','SelfConfidence','GoodPhysicalHealth']] = sc.fit_transform(X_train[['Age','Perseverance','DesireToTakeInitiative','Competitiveness','SelfReliance','StrongNeedToAchieve','SelfConfidence','GoodPhysicalHealth']])
X_test[['Age','Perseverance','DesireToTakeInitiative','Competitiveness','SelfReliance','StrongNeedToAchieve','SelfConfidence','GoodPhysicalHealth']] = sc.transform(X_test[['Age','Perseverance','DesireToTakeInitiative','Competitiveness','SelfReliance','StrongNeedToAchieve','SelfConfidence','GoodPhysicalHealth']])
```

```python
classifier1 = LogisticRegression()
classifier1.fit(X_train,y_train)
```

```python

y_pred1 = classifier1.predict(X_test)
cm = confusion_matrix(y_test, y_pred1)
print(cm)
accuracy_score(y_test, y_pred1)
```

```python
classifier2 = GaussianNB()
classifier2.fit(X_train, y_train)
```

```python
y_pred2 = classifier2.predict(X_test)
cm = confusion_matrix(y_test, y_pred2)
print(cm)
accuracy_score(y_test, y_pred2)
```

```python
classifier3 = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier3.fit(X_train, y_train)
```

```python
y_pred3 = classifier3.predict(X_test)
cm = confusion_matrix(y_test, y_pred3)
print(cm)
accuracy_score(y_test, y_pred3)
```

## Hence a maximum accuracy of 63% can be acheived by decision tree classifier
